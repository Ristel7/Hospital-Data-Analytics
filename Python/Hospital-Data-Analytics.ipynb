```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

for clean visuals

```python
sns.set(style="whitegrid", palette="Set2")
```

```python
# Load Dataset
patients = pd.read_csv("D:/InternShip/CSV/Patients.csv")
appointments = pd.read_csv("D:/InternShip/CSV/appointments.csv")
doctors = pd.read_csv("D:/InternShip/CSV/doctors.csv")
billing = pd.read_csv("D:/InternShip/CSV/billing.csv")
treatments = pd.read_csv("D:/InternShip/CSV/treatments.csv")
prescriptions = pd.read_csv("D:/InternShip/CSV/prescriptions.csv")
surgeries = pd.read_csv("D:/InternShip/CSV/surgeries.csv")
staff = pd.read_csv("D:/InternShip/CSV/staff.csv")
```

```python
# Age distribution
plt.figure(figsize=(16,6))
sns.histplot(patients['Age'], bins=20, kde=True, color="skyblue")
plt.title("Patient Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Gender Count
plt.figure(figsize=(16,6))
sns.countplot(x='Gender', data=patients, palette='pastel')
plt.title("Gender Distribution of Patients")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#  Age vs Gender
plt.figure(figsize=(16,6))
sns.boxplot(x='Gender', y='Age', data=patients, palette='Set2')
plt.title("Age Distribution by Gender")
plt.show()
```

```python
# Age Group Distribution
patients['Age_Group'] = pd.cut(patients['Age'], bins=[0,18,30,45,60,100], labels=['0-18','19-30','31-45','46-60','60+'])
plt.figure(figsize=(16,6))
sns.countplot(x='Age_Group', data=patients, palette='coolwarm')
plt.title("Patients by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python
#  Gender Ratio per Age Group
plt.figure(figsize=(16,6))
sns.countplot(x='Age_Group', hue='Gender', data=patients, palette='pastel')
plt.title("Gender Distribution Across Age Groups")
plt.show()
```

```python
#  Patients with Missing Contact Info
plt.figure(figsize=(16,4))
missing = patients[['Email', 'Phone']].isnull().sum()
missing.plot(kind='bar', color='salmon')
plt.title("Missing Contact Information Count")
plt.ylabel("Number of Missing Values")
plt.show()
```

```python
#Correlation Heatmap (Numerical Columns)
plt.figure(figsize=(16,6))
sns.heatmap(patients.select_dtypes('number').corr(), annot=True, cmap='Blues')
plt.title("Correlation Between Numerical Patient Attributes")
plt.show()
```
